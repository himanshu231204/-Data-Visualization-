# **Matplotlib**

**What is Matplotlib?**
Matplotlib is a data visualization library in Python.

It is used to create static, animated, and interactive plots.

Inspired by MATLAB's plotting style.

Commonly used with NumPy and Pandas.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

 # **1) MatPlotib Basics**

```python
x=np.linspace(0,5,11)
x 
y=x **2 
y
```

 What is plt.plot()?

plt.plot() is used to create line plots, which connect a series of points (x, y) using straight lines.



üîß Basic Syntax: 

plt.plot(x, y, format_string, **kwargs)


| Parameter       | Description                           |
| --------------- | ------------------------------------- |
| `x`             | List or array of x-axis values        |
| `y`             | List or array of y-axis values        |
| `format_string` | (Optional) Define color, marker, line |
| `**kwargs`      | Additional styling options            |


```python
plt.plot(x,y)
```

```python
plt.plot(x,y) 
plt.title(" Learning ML ") 
plt.xlabel("Progress") 
plt.ylabel("Learning")
```

## üîÅ Revision Notes ‚Äì plt.plot()

- Used to create **line plots**.
- Syntax: `plt.plot(x, y)`
- Add optional formatting:
  - Color: `'r'`, `'g'`, `'b'`, etc.
  - Marker: `'o'`, `'s'`, `'x'`, etc.
  - Line style: `'-'`, `'--'`, `':'`
- Common styling:
  - `color`, `linewidth`, `linestyle`, `marker`, `label`
- Don't forget:
  - `plt.title()`, `plt.xlabel()`, `plt.ylabel()`, `plt.grid()`, `plt.legend()`, `plt.show()`


#                    **2)Subplots In Matplotlib**

üîπ What are Subplots?

Subplots allow you to display multiple plots in a single figure (same window).

Useful for comparing multiple visualizations side by side.

üîß Syntax:

fig, axes = plt.subplots(nrows, ncols, figsize=(width, height))

| Parameter | Description                            |
| --------- | -------------------------------------- |
| `nrows`   | Number of rows of plots                |
| `ncols`   | Number of columns of plots             |
| `figsize` | Tuple to set size of the entire figure |



axes will be a 2D NumPy array of Axes objects if more than 1 row and column.

```python
plt.subplot(2,2,1)    
plt.plot(x,x)  # this will be ploted on 1strd graph
plt.subplot(2,2,4) 
plt.plot(x,y)   #this will be ploted on 4th graph

plt.subplot(2,2,3) 
plt.plot(y,x)    # this will be ploted on 3rd graph
plt.subplot(2,2,2)
plt.plot(y**2,x/2) # tis will be ploted on 2nd graph

```

## üîÅ Quick Revision ‚Äì plt.subplot()

- Use `plt.subplot(rows, cols, plot_number)`  
  ‚û§ Example: `plt.subplot(2, 2, 1)` ‚Üí 2 rows, 2 cols, 1st plot

- Indexing starts from 1 (not 0)

- You can add plots in any order  
  ‚û§ Ex: subplot(2,2,4) likhne ka matlab 4th box (bottom-right)

- Each `plt.subplot()` creates a new small plot in that position

- Use `plt.tight_layout()` to fix overlapping


# **3) Object Oriented way**

```python
fig=plt.figure()
fig.add_axes([0.1,0.1,0.8,0.8])
```

| Part                   | Meaning                                                                                                                 |
| ---------------------- | ----------------------------------------------------------------------------------------------------------------------- |
| `plt.figure()`         | Creates a blank figure (canvas)                                                                                         |
| `fig.add_axes([...])`  | Adds an axes (plot area) to the figure                                                                                  |
| `[0.1, 0.1, 0.8, 0.8]` | Is a list that defines **\[left, bottom, width, height]** of the plot ‚Äî all in **fraction of the figure size** (0 to 1) |


```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.1,0.1])
ax1.plot(x,y)
ax2=fig.add_axes([0.3,0.4,0.3,0.3])
ax2.plot(x,y)
```

| Line                                       | Description                             |
| ------------------------------------------ | --------------------------------------- |
| `fig = plt.figure()`                       | Creates a blank canvas (figure)         |
| `ax1 = fig.add_axes([0.1, 0.1, 0.1, 0.1])` | Creates a **small plot** at bottom-left |
| `ax2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])` | Creates a **larger plot** near center   |
| `ax.plot()`                                | Plotting the data inside each axes      |


## üìå fig.add_axes([left, bottom, width, height])

- Used in Object-Oriented way to manually place plot
- All values are in range **0 to 1** (fractions of full figure)
- Example:
  - `fig = plt.figure()`
  - `ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])`

- [0.1, 0.1, 0.8, 0.8] ‚Üí 10% from left and bottom, 80% width and height


## üîÅ Nested Plot using fig.add_axes()

- Use `fig.add_axes()` multiple times to make inset/nested plots
- All positions: [left, bottom, width, height] (0‚Äì1 range)
- Example:
  - `ax1 = fig.add_axes([0.1, 0.1, 0.1, 0.1])` ‚Üí Small box
  - `ax2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])` ‚Üí Bigger box
 

üß† OO vs Pyplot (Short Summary) 

| Pyplot Way (`plt.plot`) | OO Way (`ax.plot`)              |
| ----------------------- | ------------------------------- |
| Easy and quick          | More flexible and powerful      |
| Good for simple plots   | Best for multiple/complex plots |
| Uses global state       | Fully controlled object usage   |
