# **Categorical Plots**

**Categorical plots are used to visualize relationships involving categorical (discrete) data. For example, when you want to compare values across categories like gender, day, smoker/non-smoker, etc.**

**Seaborn offers several types of categorical plots through two main functions:**

**1) sns.catplot() — a figure-level function (like displot, jointplot)**

**2)Axes-level functions (used inside sns.catplot() or directly):**

**stripplot()**

**swarmplot()**

**boxplot()**

**violinplot()**

**barplot()**

**countplot()**



````python
import matplotlib.pyplot as plt 
import seaborn as sns 
````

````python
df=sns.load_dataset("tips") 
df
````

# **CountPlot**

**countplot() is used to display the count of observations in each category using bars.
It's ideal when you're dealing with categorical columns like gender, days, smokers, etc.**

Syntax: 

**sns.countplot(x='column_name', data=df)**

You can also use:

x → categorical variable on x-axis

hue → to break bars by category (e.g., gender, smoker, etc.)


````python
sns.countplot(x=df['sex'])
````

````python
sns.countplot(x=df['day'])
````

````python
sns.countplot(y=df['tip'])
````

````python
sns.countplot(x=df['sex'],hue=df['smoker'])
````

✅**What's Happening Here:** 

x=df['sex']: You're plotting the count of people by sex (Male and Female).

hue=df['smoker']: You're further breaking that down by whether each person is a smoker (Yes or No).

So, for each gender category (on the x-axis), you’ll get two bars:

One for smokers

One for non-smokers

This allows you to compare smoking habits between males and females.



````python
sns.countplot(x='day', hue='smoker',data=df)
````

# **Barplot**

**The barplot() function is used to show the relationship between a categorical variable and a continuous variable.
It displays the mean (by default) of the continuous variable for each category.**

**sns.barplot(x='category', y='value', data=df, estimator=np.function)**


 Important Parameters: 
 | Parameter | Description                                                                  |
| --------- | ---------------------------------------------------------------------------- |
| `x`       | Categorical variable (e.g., gender, day)                                     |
| `y`       | Continuous variable (e.g., tip, total\_bill)                                 |
| `hue`     | For additional categorical separation                                        |
| `est`     | Estimator function (default: `mean`, but can be `np.sum`, `np.median`, etc.) |
| `ci`      | Confidence Interval (default: 95%) — set `ci=None` to remove                 |



````python
sns.barplot(x='sex',y='tip',hue='smoker',data=df)
````

**✅ What is estimator in Seaborn?**
The estimator parameter tells Seaborn how to calculate the value to be plotted for each category.

🔹 **By default, Seaborn uses np.mean, which means it shows the average value for each category.** 

🔹 But you can change this to other functions like np.sum, np.median, len, etc., depending on what you want to analyze.

**🛠️ Commonly Used Estimator Functions:**  
| Estimator Function | What It Does                | Example Output      |
| ------------------ | --------------------------- | ------------------- |
| `np.mean`          | **Average** value (default) | Avg tip per gender  |
| `np.median`        | **Median** value            | Middle tip value    |
| `np.sum`           | **Sum** of values           | Total sales per day |
| `len`              | **Count** of observations   | Count per category  |
| `np.max`           | **Maximum** value           | Highest score       |
| `np.min`           | **Minimum** value           | Lowest salary       |



````python
import numpy as np
````

````python
sns.barplot(x='sex',y='tip',estimator=np.sum,data=df)
````

# **Boxplot**

✅ Purpose:
**boxplot() is used to visualize the distribution, median, quartiles, and outliers in a numerical column grouped by a categorical feature.**

📈 What Does It Show? 

Box: Interquartile Range (IQR: 25th to 75th percentile)

Line inside box: Median (50th percentile)

Whiskers: Data within 1.5 × IQR

Dots (fliers): Outliers beyond the whiskers

✍️ Syntax:
**sns.boxplot(x='category_column', y='numeric_column', data=df)**


````python
sns.boxplot(x='tip',y='day',data=df)
````

**🔍 Key Features:**

| Parameter    | Description                                               |
| ------------ | --------------------------------------------------------- |
| `x` or `y`   | Categorical on x-axis or y-axis, and numeric on the other |
| `data`       | Your DataFrame                                            |
| `hue`        | Adds sub-categories (second categorical variable)         |
| `palette`    | Change the color theme (`'Set2'`, `'coolwarm'`, etc.)     |
| `orient`     | Orientation (`'h'` for horizontal, `'v'` for vertical)    |
| `width`      | Width of the box                                          |
| `linewidth`  | Border thickness of the box                               |
| `showfliers` | Show or hide outliers (`True` or `False`)                 |


````python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
````

# **Violinplot**

**A violin plot combines the features of a boxplot and a KDE (kernel density estimate) plot.
It not only shows summary statistics (like median, quartiles), but also the distribution shape of the data.**

✅ Syntax: 

**sns.violinplot(x='category_column', y='numerical_column', data=df)** 

🔧 Key Parameters:

| Parameter    | Purpose                                                                                |
| ------------ | -------------------------------------------------------------------------------------- |
| `x` / `y`    | Categorical and numerical axes                                                         |
| `data`       | DataFrame to plot                                                                      |
| `hue`        | Used to group data within each category                                                |
| `palette`    | Controls color theme                                                                   |
| `split=True` | When used with `hue`, it splits violins in half                                        |
| `inner`      | Controls what's inside the violin: `'box'`, `'quartile'`, `'point'`, `'stick'`, `None` |
| `scale`      | `'area'` (default), `'count'`, or `'width'` — affects width of violin                  |


````python
sns.violinplot(x='day',y='total_bill',hue='sex',data=df)
````

# **Stripplot**

📍**What is stripplot()?**

A strip plot is used to plot individual data points along a categorical axis.

It’s like a scatter plot but for categorical variables.

Great for visualizing the distribution of a small dataset.

✅ Syntax: 

**sns.stripplot(x='category_column', y='numerical_column', data=df)** 

🔧 Key Parameters: 

| Parameter | Description                                               |
| --------- | --------------------------------------------------------- |
| `x` / `y` | Categorical and numerical columns                         |
| `data`    | DataFrame containing the data                             |
| `hue`     | Grouping variable for subcategories                       |
| `jitter`  | Adds random noise for better visibility (default: `True`) |
| `dodge`   | Separates points when using `hue`                         |
| `palette` | Color theme                                               |
| `size`    | Marker size                                               |
| `alpha`   | Transparency of the points (range: 0 to 1)                |


````python
sns.stripplot(x='day',y='total_bill',hue='sex',data=df)
````

# **Swarmplot**

🐝 **What is swarmplot()?**

A swarm plot is like a strip plot, but it automatically adjusts the position of the points to avoid overlaps.

It provides a clearer picture of each individual observation, especially in dense areas.

It is more precise than stripplot() in avoiding overlaps.

✅ Syntax: 

**sns.swarmplot(x='category_column', y='numerical_column', data=df)**

🔧 Key Parameters:

| Parameter | Description                              |
| --------- | ---------------------------------------- |
| `x` / `y` | Categorical and numerical columns        |
| `data`    | DataFrame containing the data            |
| `hue`     | Categorical variable for color grouping  |
| `dodge`   | Separate hue groups horizontally         |
| `palette` | Color style (e.g., `'Set2'`, `'pastel'`) |
| `size`    | Size of the markers                      |
| `alpha`   | Transparency of points                   |


````python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=df)
````

## You can combine violinplot() and swarmplot() in Seaborn to get both:

**The distribution of the data (from violinplot)**

**The individual data points (from swarmplot)**

**This combo gives a powerful visual!**

````python
sns.violinplot(x='day',y='total_bill',hue='sex',data=df) 
sns.swarmplot(x='day',y='total_bill',hue='sex',data=df)
````

# 📊 Categorical Plots in Seaborn – Summary

1️⃣ **Countplot**  
👉 Shows the **count** (frequency) of each category.

2️⃣ **Barplot**  
👉 Shows the **mean** of a numerical value for each category.

3️⃣ **Boxplot**  
👉 Displays **distribution**, **median**, **quartiles**, and **outliers**.

4️⃣ **Violinplot**  
👉 Combines **boxplot** with a **KDE (density)** curve to show distribution shape.

5️⃣ **Stripplot**  
👉 Plots **individual data points** with slight jitter (may overlap).

6️⃣ **Swarmplot**  
👉 Like stripplot but **avoids overlapping points**.

7️⃣ **Combination (Violin + Swarm)**  
👉 Violin for distribution + swarm for exact data points.
