# **Regression Plot**

**A regression plot is used to visualize the relationship between two continuous variables, and includes a best-fit line (regression line) that helps us understand the trend.**

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
tips=sns.load_dataset('tips') 
tips
```

✅ **lmplot() – What It Does:**

lmplot() is used to plot linear relationships between two variables (just like regplot) but offers additional flexibility like:

Grouping data with hue

Creating faceted grids with col or row

Styling with markers, palette, etc.



```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips)
```
